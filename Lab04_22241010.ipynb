```python
class Patient:
    def __init__(self, id = None, name = None, age = None, bloodgroup = None):
        self.prev = None
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = None

class WRM:
    def __init__(self):
        self.head = None
        self.tail = None

    def __dummy_head_creation(self): #*private func of WRM class
        self.head = Patient()
        self.head.prev = self.tail
        self.head.next = self.tail

    def RegisterPatient(self, id, name, age, bloodgroup):
        if self.head == None or self.head.next == None:
            self.tail = Patient(id, name, age, bloodgroup)
            self.__dummy_head_creation()
            self.tail.prev = self.head
            self.tail.next = self.head
            return

        node = Patient(id, name, age, bloodgroup)
        self.tail.next = node
        self.head.prev = node
        node.prev = self.tail
        node.next = self.head
        self.tail = node

    def ServePatient(self):
        if self.CanDoctorGoHome():
            print("No more patients to serve")
            return
        node = self.head.next
        self.head.next = node.next
        node.next.prev = self.head
        if node == self.tail:
            self.CancelAll()
        return node.name

    def ShowAllPatient(self):
        if self.CanDoctorGoHome():
            print("No more patients to serve")
            return
        current = self.head.next
        main_str = ""
        while current != self.head:
            main_str += str(current.name) + " "
            current = current.next
        return main_str

    def CanDoctorGoHome(self):
        if self.head == None:
            return True
        elif self.head.next == None:
            return True
        return False

    def CancelAll(self):
        if self.tail != None:
            self.tail.prev = None
            self.tail.next = None
            self.tail = None
        if self.head != None:
            self.head.next = None
            self.head.prev = None

    def ReverseTheLine(self):
        if self.head == None or self.tail == None:
            return
        current = self.head.prev #previous item of head
        self.tail = self.head.next #setting the tail to next item of head
        while current != self.head:
            node = current.prev #storing the previous item of current
            current.next.next = current #setting the next item of current.next as current
            current.prev = current.next #setting the previous item of current.prev as current
            current = node
        current.prev = self.tail #at the end setting dummy head prev to tail
        self.tail.next = current #the tail's next item is dummy head

    def run_commands(self, num):
        if num == 1:
            id = input("Enter ID: ")
            name = input("Enter name: ")
            age  = input("Enter age: ")
            blood_grp = input("Enter blood group: ")
            self.RegisterPatient(id, name, age, blood_grp)
            print("Success Registering Patient")
        elif num == 2:
            p_served = self.ServePatient()
            if p_served:
                print(f"{p_served} is served")
        elif num == 3:
            self.CancelAll()
            print("All appointments cancelled")
        elif num == 4:
            if self.CanDoctorGoHome():
                print("Yes")
            else:
                print("No")
        elif num == 5:
            print(self.ShowAllPatient())
        elif num == 6:
            self.ReverseTheLine()
            print("The patient line has been reversed")

if __name__ == '__main__':
    print("Welcome to Waiting Room Management System**")
    wrm = WRM()
    #* Tester Code
    while True:
        print("**Choose an option**")
        func_lst = ["RegisterPatient()", "ServePatient()", "CancelAll()",
               "CanDoctorGoHome()", "ShowAllPatients()", "ReverseTheLine()","exit"]

        for i, ele in enumerate(func_lst):
            print(f"{i + 1}. {ele}")
        print("******************")
        user = None

        while True:
            user = input("Enter your choice: ")
            try:
                user = int(user)
            except:
                print("Invalid choice, input a number")
            else:
                if user < 1 or user > 7:
                    print("Please enter a number from 1 to 7")
                else:
                    break
        if user == 7:
            print("Thank You For Using Waiting Room Management System")
            print("Exiting...")
            break
        else:
            print(f"executing {func_lst[user - 1]}...")
            print("")
            wrm.run_commands(user)
            print("")
```
